```python
# 2. Imports
import pandas as pd
from datasets import load_dataset
from transformers import pipeline
import re
```

```python
# 3. Load Dataset (Spider subset)
dataset = load_dataset("spider", split="train[:50]")

data = []
for item in dataset:
    data.append({
        "question": item["question"],
        "gold_sql": item["query"]
    })

df = pd.DataFrame(data)

print("Dataset size:", df.shape)
df.head()
```

```python
# 4. Load Models
model_1 = pipeline(
    "text-generation",
    model="google/flan-t5-base",
    max_length=128
)

model_2 = pipeline(
    "text-generation",
    model="tiiuae/falcon-rw-1b",
    max_length=128
)
```

```python
# 5. Zero-Shot Prompt
def generate_sql(model, question):
    prompt = f"""
Convert the following natural language question into SQL.
Only output a valid SQL query starting with SELECT.

Question: {question}
SQL:
"""

    output = model(prompt, do_sample=True, temperature=0.7)
    text = output[0]["generated_text"]

    # Extract only SELECT query
    import re
    match = re.search(r"(SELECT .*?;)", text, re.IGNORECASE)

    if match:
        return match.group(1)
    return text
```

```python
# 6. Generate Predictions
df["model_1_sql"] = df["question"].apply(lambda q: generate_sql(model_1, q))
df["model_2_sql"] = df["question"].apply(lambda q: generate_sql(model_2, q))

df.head()
```

```python
# 7. Evaluation
def normalize(sql):
    sql = sql.lower()
    sql = re.sub(r"\s+", " ", sql)
    return sql.strip()
```

```python
# Exact Match
def exact_match(pred, gold):
    return normalize(pred) == normalize(gold)
```

```python
# Soft Match
def soft_match(pred, gold):
    pred = normalize(pred)
    gold = normalize(gold)

    return (
        pred in gold or
        gold in pred or
        any(word in pred for word in gold.split())
    )
```

```python
# Execution Validity
def execution_valid(sql):
    sql = normalize(sql)
    return (
        "select" in sql and
        ("from" in sql or "where" in sql)
    )
```

```python
# 8. Compute Metrics
results = []

for name, col in [("Flan-T5", "model_1_sql"), ("Falcon", "model_2_sql")]:

    exact = df.apply(lambda x: exact_match(x[col], x["gold_sql"]), axis=1).mean() * 100
    soft = df.apply(lambda x: soft_match(x[col], x["gold_sql"]), axis=1).mean() * 100
    valid = df[col].apply(execution_valid).mean() * 100

    results.append({
        "Model": name,
        "Exact Match (%)": round(exact, 2),
        "Soft Match (%)": round(soft, 2),
        "Execution Validity (%)": round(valid, 2)
    })

results_df = pd.DataFrame(results)
results_df
```

```python
# 9. Save Results
results_df.to_csv("final_results.csv", index=False)
print("Saved as final_results.csv")
```
